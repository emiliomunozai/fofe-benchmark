#### Complete data processed

```python
from src.dataloaders import build_corpus_examples

examples, vocab = build_corpus_examples(
    split="train",
    max_docs=500000,          
    max_seq_len=4096,    
    min_distance=32,
    max_distance=2048,
)

# Save examples and vocab into data folder
import pickle

os.makedirs(os.path.join(os.getcwd(), "data"), exist_ok=True)
examples_path = os.path.join(os.getcwd(), "data", "examples.pkl")
vocab_path = os.path.join(os.getcwd(), "data", "vocab.pkl")

with open(examples_path, "wb") as f:
    pickle.dump(examples, f)

with open(vocab_path, "wb") as f:
    pickle.dump(vocab, f)

print(f"Saved {len(examples)} examples to {examples_path}")
print(f"Saved vocab (size: {len(vocab)}) to {vocab_path}")
```

```python
visualize_examples(examples, vocab)
```
